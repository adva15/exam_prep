```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/pythonai250824/exam_prep/main/mpg.csv")
df
```

```python
# תשובה לשאלה 1
sns.scatterplot(data=df, x='horsepower', y='mpg', hue='origin', palette='Spectral')

plt.show()
```

```python
# תשובה לשאלה 2
Consume_less_fuel4_or8 = df[df['cylinders'].isin([4, 8])]
sns.barplot(data=Consume_less_fuel4_or8, x='cylinders', y='mpg', hue='cylinders', palette='Spectral', estimator=np.mean)

plt.show()
```

```python
# תשובה לשאלה 3
sns.countplot(data=df, x='origin', hue='origin', palette='Spectral')
plt.show()
```

```python
# תשובה לשאלה 4
sns.histplot(data=df, x='weight',bins=20, kde=True)
plt.show()
```

```python
# תשובה לשאלה 5
plt.figure(figsize=(10, 4))
sns.boxplot(data=df, x='cylinders', y='mpg',hue='cylinders', palette='Spectral')
plt.show
```
